```python
#실습03

import numpy as np
import pandas as pd

titan = pd.read_csv('titanic_train.csv')
titan['Age'].fillna(titan['Age'].mean(), inplace=True)
titan['Cabin'].fillna('N', inplace=True)
titan['Embarked'].fillna('N', inplace=True)
```

```python
titan
```

```python
titan.groupby('Sex')['Survived'].sum()
```

```python
titan.groupby('Pclass')['Survived'].sum()
```

```python
def group_age(n):
    n = int(n)
    if n // 10 == 0:
        return "유아"
    elif n // 10 >= 7:
        return "노인"
    else:
        return str(n // 10) + "0대"
```

```python
group_age(29)
```

```python
titan["Age_group"] = titan["Age"].apply(group_age)
```

```python
titan
```

```python
titan.groupby(['Age_group'])['Survived'].sum()
```

```python
titan.to_csv('titan_save.csv',encoding='utf-8-sig')
```
