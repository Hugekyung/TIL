```
import scrapy
import pandas as pd

class NewsSpider(scrapy.Spider):
    name = 'newsbot'

    start_urls = ['https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=101&date=20200822']

    def parse(self, response):
        titles = response.css('a::attr(title)').extract()
        offices = response.css('div.ranking_office::text').extract()
        previews = response.css('div.ranking_lede::text').extract()

        for item in zip(titles, offices, previews):
            scraped_info = {
                            'title' : item[0].strip(),
                            'office' : item[1].strip(),
                            'preview' : item[2].strip()
            }
            
            yield scraped_info
```
