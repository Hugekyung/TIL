```python
import pandas as pd

df = pd.read_csv(r'V4.csv')
com = list(df['comment'])
review = ', '.join(com)
```

```python
df
```

```python
from collections import Counter
from konlpy.tag import Okt

okt = Okt()
noun = okt.nouns(review)
for i,v in enumerate(noun):
    if v == '브이포' or v == '게임' or len(v) < 2 or v == '넥슨' or v == 'v4':
        noun.pop(i)

count = Counter(noun)

noun_lst = count.most_common(100)
for i in noun_lst:
    print(i)
```
